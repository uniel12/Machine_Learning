## 데이터 셋 구성하기
- 문제(X)와 답(y) 분리하기
- 훈련세트와 평가세트 부리
    - 훈련세트/훈련데이터/train : 머신러닝 모델을 학습할 때 사용
    - 평가세트/평가데이터/홀드아웃 세트 : 모델이 얼마나 잘 작동하는지 측정하는데 사용

```python
import pandas as pd
```

```python
# 문제데이터 생성 -> 데이터프레임 형태로 생성
# data 키값을 활용해서 데이터프레임 생성
# columns 값은 feature_names 활용
# pd.DataFrame(데이터(value), columns =)
iris_df = pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
iris_df
```

```python
# 문제와 답 데이터로 분리
# 문제 데이터는 대문자 X에 담아주기(2차원)
# 답 데이터는 소문자 y에 담아주기(1차원) <- 따로 시리즈 형태로 만들어주지 않아도 됨 차원이 중요!
X = iris_df
y = iris_data['target']
```

```python
from sklearn.model_selection import train_test_split
# 훈련세트와 평가세트로 분리하기
# train_test_split 함수 : 훈련세트와 테스트 세트를 분리해주는 기능
# 데이터를 나누기 전에 데이터 셋을 무작위로 섞어준다.
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3, # default = 0.25
                                                    random_state=65) 
# random_state: 함수를 여러번 실행해도 결과가 똑같이 나온다.
# 데이터 분할시 셔플이 이루어지는데 이를 위한 시드값 (int나 RandomState로 입력)
```

```python
# X_train, X_test, y_train, y_test 크기/ 모양 = shape

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### EDA (데이터 살펴보기)

```python
# pd.plotting.scatter_matrix 함수 : 데이터프레임 산점도 행렬 생성
pd.plotting.scatter_matrix(iris_df,
                           figsize = (15,15), # 행렬의 크기
                           marker = '^', # 그래프의 점 모양
                           c = y, # 산점도 그래프의 점 색상 / y는 클래스레이블(3개 0,1,2)
                           alpha = 0.8) #alpha는 점의 투명도 1로 갈수록 불투명
```

- 세 클래스가 꽃잎과 꽃받침의 측정값에 따라 비교적 잘 구분됨
- 머신러닝모델을 학습시키는데 적합하다고 판단됨

### 모델링

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
# 모델 생성 - knn이라는 변수에 담아주기
knn = KNeighborsClassifier(n_neighbors=3)
knn
```

```python
# 모델 학습 fit()메서드 사용
knn.fit(X_train,y_train)
```

```python
# 예측하기 predict()메서드 사용
pre = knn.predict(X_test)
pre
```

```python
# 평가하기 metrics 함수 사용
metrics.accuracy_score(pre,y_test) # 정확도
```

### 하이퍼파라미터 튜닝

```python
test_list = []  #평가데이터를 통한 성능 평가값이 담긴 리스트
train_list = [] #훈련데이터를 통한 성능 평가값 담긴 리스트

for k in range(1,105) :
    # 모델생성
    knn = KNeighborsClassifier(n_neighbors=k)
    # 모델학습
    knn.fit(X_train,y_train)
    # 예측
    test_pre = knn.predict(X_test)
    # test 평가(정확도)
    test_result = metrics.accuracy_score(test_pre,y_test)
    # test_list라는 리스트에 결과값을 저장
    test_list.append(test_result)
    
    # train 예측
    train_pre = knn.predict(X_train) 
    # train 평가(정확도)
    train_result = metrics.accuracy_score(train_pre,y_train)
    # 평가값을 train_list 결과값을 저장
    train_list.append(train_result)
```

```python
test_list
```

```python
train_list
```

```python
# 훈련용/테스트 결과 시각화하기
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
# train_list 그래프 그리기
plt.plot(range(1,105),train_list, label="train") # label값 달아주기
# test_list 그래프 그리기
plt.plot(range(1,105),test_list, label="test")
# 범례표시
plt.legend()

plt.show()
```
