## 로지스틱 회귀 (Logistic Regression)
- 선형회귀 : 이상치에 너무 민감!! => 분류를 진행하지 않는다
- 로지스틱 회귀 등장!
- 시그모이드 함수 사용
- x -> 무한대 가정시 1 / x -> -무한대 가정시 0
- 리턴값 0과 1 사이의 연속적인 값 0.5기준으로 0과 1로 분류

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
iris_data = load_iris()
iris_data
```

```python
# DESCR 키 데이터셋에 대한 설명
# petal : 꽃잎 / sepal : 꽃받침
print(iris_data.DESCR)
```

```python
# 문제 데이터 -> 데이터 프레임 형태로 변환
X = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
X
```

```python
# 답데이터 -> 데이터 프레임 형태로 변환
y = pd.DataFrame(iris_data.target,columns=['class'])
y
```

```python
# 훈련세트, 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=65) 
```

```python
# 모델 생성
logi_model = LogisticRegression(solver='saga', # 'saga' : L1 정규화 진행
                                max_iter=2000) # max_iter : 몇번 학습을 진행할건지
```

```python
# 모델 학습
logi_model.fit(X_train,y_train)
```

```python
# 예측
pre = logistic_model.predict(X_test)
```

```python
# 결과확인
accuracy_score(pre,y_test)
```
